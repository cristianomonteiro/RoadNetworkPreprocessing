# Loading the data

```python
params = {'user': 'cristiano', 'password': 'cristiano'}
kdf = ks.read_sql_query('''	select	EDGE.IDVERTEXORIG_FK,
									EDGE.IDVERTEXDEST_FK,
									EDGE.LENGTH
							from	BACKUP_EDGES as EDGE ''',
        		'jdbc:postgresql:afterqualifying', options=params)

G = nx.from_pandas_edgelist(kdf.to_pandas(), 'idvertexorig_fk', 'idvertexdest_fk', ['length'])
kdf = None
```

# Comparing the smaller components

```python
components = sorted(nx.connected_components(G), key=len, reverse=True)

for comp in components:
	print(str(len(comp)) + ' ' + str(list(comp)[0]))
```
